## 2. Estimación de la densidad por kernel

#### a) KDEUnivariate de statsmodels

```python
from statsmodels.nonparametric.kde import KDEUnivariate

kde1= KDEUnivariate(arrDatos)
kde1.fit(bw=0.2736)

plt.figure(figsize=(9,4))
plt.plot(kde1.support, [kde1.evaluate(xi) for xi in kde1.support],'-')

# kde1.support son los puntos de eje X
```

#### b) KDEMultivariate de statsmodels

```python
from statsmodels.nonparametric.kernel_density import KDEMultivariate

# x_grid son los puntos de eje X
x_grid = np.linspace(2, 10, 1000)

kde = KDEMultivariate(arrDatos, bw=0.2736 * np.ones_like(arrDatos), var_type='c')
pdf = kde.pdf(x_grid)

plt.figure(figsize=(9,4))
plt.plot(x_grid, pdf, color='blue', alpha=0.5, lw=3)
```

#### b) KernelDensity de Scikit-learn

```python
from sklearn.neighbors import KernelDensity

# x_grid son los puntos de eje X
x_grid = np.linspace(2, 10, 1000)

# eje Y
kde_skl = KernelDensity(bandwidth=0.2736)
kde_skl.fit(arrDatos[:, np.newaxis])

# score_samples() returns the log-likelihood of the samples
log_pdf = kde_skl.score_samples(x_grid[:, np.newaxis])
pdf = np.exp(log_pdf)

plt.figure(figsize=(9,4))
plt.plot(x_grid, pdf, color='blue', alpha=0.5, lw=1.5) # alpha = transparencia, lw = grosor linea
plt.fill(x_grid, pdf, ec='gray', fc='gray', alpha=0.3)
```

### Seleccion del ancho de banda
Vamos a llamar al metodo varias veces, una por cada ancho de banda y lo mejor es crear una funcion para no repetir la llamada en el codigo tantas veces como anchos diferentes tengamos.

```python
def kde_sklearn(x, x_grid, bandwidth=0.2, **kwargs):
    kde_skl = KernelDensity(bandwidth=bandwidth, **kwargs)
    kde_skl.fit(x[:, np.newaxis])
    
    # score_samples() returns the log-likelihood of the samples
    log_pdf = kde_skl.score_samples(x_grid[:, np.newaxis])
    return np.exp(log_pdf)
```

```python
fig, ax = plt.subplots()
fig.figsize=(9,4)

for bandwidth in [0.1, 0.3, 1.0]:
    ax.plot(x_grid, 
            kde_sklearn(arrDatos, x_grid, bandwidth=bandwidth), 
            label='bw={0}'.format(bandwidth), 
            linewidth=2, 
            alpha=0.5)
    
ax.hist(arrDatos, 25, fc='gray', histtype='stepfilled', alpha=0.5, density=True)
ax.set_xlim(3.5, 8.5)
ax.legend(loc='upper left')
```

Vemos que el ancho de banda afecta mucho a la forma de la funcion de densidad.
Para elegir el ancho correcto tenemos dos estrategias:
1. Usar reglas de referencia (preferido por los estadistas)
2. Usar validacion cruzada (preferido por el mundo del ML)

Because the various reference rules generally depend on (often dubious) assumptions about the underlying distribution of the data, bandwidth selection based in cross-validation can produce more trustworthy results for real-world datasets. 

#### Ancho de banda por validacion cruzada en Scikit-Learn
En Scikit-Learn es muy simple usando el estimador <code style="font-family: Arial; background:#d4d9de;color:black;font-size:1.0em ">GridSearchCV</code>.

```python
from sklearn.grid_search import GridSearchCV

grid = GridSearchCV(KernelDensity(),
                    {'bandwidth': np.linspace(0.1, 1.0, 30)},
                    cv=20) # 20-fold cross-validation
grid.fit(arrDatos[:, None])
print (grid.best_params_)
```

#### Kernels

```python
def plot_kernels():
    #Visualize the KDE kernels available in Scikit-learn"""
    fig, ax = plt.subplots(figsize=(8, 6),
                           #subplot_kw={'axisbg':'#EEEEEE', 'axisbelow':True}
                          )
    ax.grid(color='white', linestyle='-', linewidth=2)
    for spine in ax.spines.values():
        spine.set_color('#BBBBBB')

    X_src = np.zeros((1, 1))
    x_grid = np.linspace(-3, 3, 1000)

    for kernel in ['gaussian', 'tophat', 'epanechnikov',
                   'exponential', 'linear', 'cosine']:
        log_dens = KernelDensity(kernel=kernel).fit(X_src).score_samples(x_grid[:, None])
        ax.plot(x_grid, np.exp(log_dens), lw=3, alpha=0.5, label=kernel)
    ax.set_ylim(0, 1.05)
    ax.set_xlim(-2.9, 2.9)
    ax.legend()
```

```python
plot_kernels()
```
